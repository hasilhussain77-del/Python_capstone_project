Task 2 - The cost column in the dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use the “For” loop for this task.

```python
# starting with sum = 0 and count = 0
# these will help us calculate the running average
cost_sum = 0
count = 0

# loop through each row of the dataframe
for i in range(len(project_data)):
    
    # check if the Cost value is missing
    if pd.isna(project_data.loc[i, "Cost"]):
        
        # calculate running average using previous values
        running_avg = cost_sum / count
        
        # replace the missing value with running average
        project_data.loc[i, "Cost"] = running_avg
        
        # add this value to the sum and increase count
        cost_sum += running_avg
        count += 1
        
    else:
        # if cost value already exists
        # add it to running sum
        cost_sum += project_data.loc[i, "Cost"]
        
        # increase the count
        count += 1


# remove extra column "cost" if it exists
if "cost" in project_data.columns:
    project_data = project_data.drop(columns=["Cost"])

# display the dataframe
project_data

```

Task 3
Split the name column in the Employee dataframe into two new columns “First Name”,
and “LastName” and remove the older “name” column.

```python
# split the Name column into two parts using space
# expand=True creates two separate columns
Employee_data[["First Name","LastName"]] = Employee_data["Name"].str.split(" ", expand=True)

# remove the old Name column since we already created first and last name
Employee_data = Employee_data.drop(columns=["Name"])

# display the dataframe
Employee_data
```

Task 4
Join all three dataframes in one single dataframe. Name it “Final”

```python
# first we need to merge the project data and employee data on the basis of id.
# this will combine project details with employee details
Final = pd.merge(project_data, Employee_data, on="ID")

# now we will merge the result with third table that is seniority level using id column.
# this will add designation level for each employee
Final = pd.merge(Final, Seniority_level, on="ID")

# display the final combined dataframe
Final
```

Task 5
Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
cost only to employees who have finished the projects.

```python
# we need to create a bonus column and give value 0 
Final["Bonus"] = 0

# for loop is created to go through each row and identify the status is finished or not.
for i in range(len(Final)):
    
    # this loop checks if the status is finished.
    if Final.loc[i, "Status"] == "Finished":
        
        # at last we will multiply by 0.05 to get the final bonus amount.
        Final.loc[i, "Bonus"] = Final.loc[i, "Cost"] * 0.05

# display the final dataframe
Final
```

Task 6
Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.


```python
# using for loop to check each row
for i in range(len(Final)):
    
    # checking if the status is failed or not.
    if Final.loc[i, "Status"] == "Failed":
        
        # time to reduce the designation level by 1 for failed projects
        Final.loc[i, "Designation Level"] = Final.loc[i, "Designation Level"] - 1


# removeing the employees whose designation level is more than 4
Final = Final[Final["Designation Level"] <= 4]

# display the updated dataframe
Final
```

Task 7
Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# creating for loop to go through each row
for i in range(len(Final)):
    
    # checking if the gender is male
    if Final.loc[i, "Gender"] == "M":
        
        # if its male add mr
        Final.loc[i, "First Name"] = "Mr. " + Final.loc[i, "First Name"]
    
    else:
        # iif its female add mrs
        Final.loc[i, "First Name"] = "Mrs. " + Final.loc[i, "First Name"]


# at last we need to remove gender column becuase its not needed
Final = Final.drop(columns=["Gender"])

# display the dataframe
Final
```

Task 8
Promote designation level by 1 for the employees whose age is more than 29 years using
IF condition.

```python
# going through all row using for loop
for i in range(len(Final)):
    
    # checking if the age is more than 29
    if Final.loc[i, "Age"] > 29:
        
        # Add 1 for each designation level if age is more than 29
        Final.loc[i, "Designation Level"] = Final.loc[i, "Designation Level"] + 1


# display the updated dataframe
Final
```

Task 9
Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
# using group by we can add the cost of each employee's project
# we need to rest index so that grouped column will not become index.
TotalProjCost = Final.groupby(["ID","First Name"])["Cost"].sum().reset_index()

# rename the Cost column to Total cost
TotalProjCost = TotalProjCost.rename(columns={"Cost":"Total cost"})

# display the new dataframe
TotalProjCost
```
